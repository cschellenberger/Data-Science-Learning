### Keys in a dictionary need to be immutable. Therefore lists are not compatible with dictionaries because we can change the elements of a list.

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }

# Add italy to europe
europe['italy'] = 'rome'

# Print out italy in europe
print('italy' in europe)

# Add poland to europe
europe['poland'] = 'warsaw'

# Print europe
print(europe)
```

### Dictionaries can contain key:value pairs where the values are again dictionaries. These internal dictionaries are called #Sub-dictionaries

europe['spain']['population']

   - Column access — df[['column_ref_1', 'column_ref_2']]
   - Row access: only through slicing — df[1:4]
- loc (label-based)
   - Row access — df.loc[['row_label_1', 'row_label_2']]
   - Column access — df.loc[:, ['column_ref_1', 'column_ref_2']]
   - Row&Column access — df.loc[['row_label_1', 'row_label_2'], ['column_ref_1', 'column_ref_2']]

   - Row access — df.iloc[[1,2,3]]
   - Column&Row access — df.iloc[[1,2,3], [0,1]]
     - df.iloc[:, [0,1]]
