### Keys in a dictionary need to be immutable. Therefore lists are not compatible with dictionaries because we can change the elements of a list.

dictionary['key'] will output the value in the dictionary associated with that key.

'key' in dictionary will output True or False depending on whether or not the key exists within the dictionary

````python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }

# Add italy to europe
europe['italy'] = 'rome'

# Print out italy in europe
print('italy' in europe)

# Add poland to europe
europe['poland'] = 'warsaw'

# Print europe
print(europe)
````

### Using loc and iloc to call data from your Pandas dataframe:
- square brackets
   - Column access — 
   ``` python
   df[['column_ref_1', 'column_ref_2']]
   ```
   - Row access: only through slicing — 
   ``` python
   df[1:4]
   ```
- loc (label-based)
   - Row access — 
   ``` python
   df.loc[['row_label_1', 'row_label_2']]
   ```
   - Column access — 
   ``` python
   df.loc[:, ['column_ref_1', 'column_ref_2']]
   ```
   - Row&Column access — 
   ``` python
   df.loc[['row_label_1', 'row_label_2'], ['column_ref_1', 'column_ref_2']]
   ```
- iloc (index-based)
   - Row access — 
   ``` python
   df.iloc[[1,2,3]]
   ```
   - Column&Row access — 
   ``` python
   df.iloc[[1,2,3], [0,1]]
   ```
   
     ``` python
     df.iloc[:, [0,1]]
     ```
